```python
import numpy as np
import pandas as pd

from open_prices.config import EXTERNAL_DATA_FILE
from open_prices.dataset import checkListTypeAndConvert, noneSumCalc

pd.options.display.float_format = '{:.2f}'.format
```

```python
df = pd.read_parquet(f"{EXTERNAL_DATA_FILE}", engine="fastparquet")
```

```python
df.head()
```

```python
dfNone = noneSumCalc(df)
```

```python
dfNone.sort_values('noneSum', ascending=False)
```

```python
removeColumns = dfNone.sort_values('noneSum', ascending=False)[0:2]["columns"].values
```

```python
df['store_name'] = df['location_osm_display_name'].str.split(',', expand=True)[0]
```

```python
removeColumns = np.append(
    removeColumns,
    [
        'origins_tags',
        'proof_source',
        'proof_created',
        'proof_updated',
        'location_id',
        'proof_file_path',
        'proof_mimetype',
        'location_created',
        'location_updated',
        'location_osm_display_name',
    ],
)
```

```python
removeColumns
```

```python
newDF = df.drop(axis=1, columns=removeColumns)
```

```python
newdfNone = noneSumCalc(newDF)
```

```python
newdfNone.sort_values('noneSum', ascending=False)
```

```python
checkListTypeAndConvert(newDF, True)
```

```python
newDF["category_tag"] = newDF["category_tag"].replace(to_replace='[a-z]{2}:', regex=True, value='')
newDF["labels_tags"] = newDF["labels_tags"].str.replace(r"^\[\'\w{2}:(\w+)\']$", r"\1", regex=True)
```
